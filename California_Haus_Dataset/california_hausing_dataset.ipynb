```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('housing_sample_1000.csv')
```

```python
df
```

```python
#----Not Scaled----

plt.plot(df['MedHouseVal'], label='MedHouseVal', color='blue')
plt.plot(df['AveRooms'], label='AveRooms', color='yellow')
plt.plot(df['MedInc'], label='MedInc', color='red')
plt.plot(df['HouseAge'], label='HouseAge', color='green')
plt.legend()
```

```python
plt.plot(df['MedHouseVal'], label='MedHouseVal', color='blue')
plt.plot(df['AveBedrms'], label='AveBedrms', color='red')
plt.plot(df['Population'], label='Population', color='green')
plt.plot(df['AveOccup'], label='AveOccup', color='yellow')
plt.legend()
```

```python
#----Scaled----
scaler = StandardScaler()
MedInc_scaled = scaler.fit_transform(np.array(df['MedInc']).reshape(-1,1))
MedHouseVal_scaled = scaler.fit_transform(np.array(df['MedHouseVal']).reshape(-1,1))

plt.plot(MedInc_scaled, label='MedInc', color='red')
plt.plot(MedHouseVal_scaled, label='MedHouseVal', color='green')
plt.legend()
df[['MedInc','MedHouseVal']].corr()
```

```python
scaler = StandardScaler()
HouseAge_scaled = scaler.fit_transform(np.array(df['HouseAge']).reshape(-1,1))
MedHouseVal_scaled = scaler.fit_transform(np.array(df['MedHouseVal']).reshape(-1,1))

plt.plot(HouseAge_scaled, label='HouseAge', color='red')
plt.plot(MedHouseVal_scaled, label='MedHouseVal', color='green')
plt.legend()
df[['HouseAge','MedHouseVal']].corr()
```

```python
scaler = StandardScaler()
AveRooms_scaled = scaler.fit_transform(np.array(df['AveRooms']).reshape(-1,1))
MedHouseVal_scaled = scaler.fit_transform(np.array(df['MedHouseVal']).reshape(-1,1))

plt.plot(AveRooms_scaled, label='AveRooms', color='red')
plt.plot(MedHouseVal_scaled, label='MedHouseVal', color='green')
plt.legend()
df[['AveRooms','MedHouseVal']].corr()
```

```python
scaler = StandardScaler()
AveBedrms_scaled = scaler.fit_transform(np.array(df['AveBedrms']).reshape(-1,1))
MedHouseVal_scaled = scaler.fit_transform(np.array(df['MedHouseVal']).reshape(-1,1))

plt.plot(AveBedrms_scaled, label='AveBedrms', color='red')
plt.plot(MedHouseVal_scaled, label='MedHouseVal', color='green')
plt.legend()

df[['AveBedrms','MedHouseVal']].corr()
```

```python
scaler = StandardScaler()
Population_scaled = scaler.fit_transform(np.array(df['Population']).reshape(-1,1))
MedHouseVal_scaled = scaler.fit_transform(np.array(df['MedHouseVal']).reshape(-1,1))

plt.plot(Population_scaled, label='Population', color='red')
plt.plot(MedHouseVal_scaled, label='MedHouseVal', color='green')
plt.legend()
df[['Population','MedHouseVal']].corr()
```
